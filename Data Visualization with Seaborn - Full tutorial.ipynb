# Relational plots

Relational plots are plots that show relationship between two quantitative variables.

Common relational plots include:

* scatter plots
* line plots

These plots can be plotted with low level functions or  high level functions which has more flexibility in seaborn.

Let's look at demonstrations with the tips data.


```python
tips.head()
```

### Scatter plot

```python
# creating a scatterplot with a low level function - sactterplot() 

sns.scatterplot(x='total_bill',y='tip',data = tips)
plt.show()
```

### lineplot
Lines can be used to show relationship or trend but is more meaningful when used to show trend.


```python
# creating a lineplot with a low level function - lineplot() 
sns.lineplot(x='total_bill',y='tip',data = tips)
plt.show()
```

### Creating relational plots with high level function: relplot()

####  relplot (Relational plot)

relplot() is for creating plots that shows relationship between two quantitatitative variables.

It is very flexible and allows subplotting. replot() provides 'scatter' and 'line' as optional plots to choose.

It allows you to create subplots (row and col) and subgroups with color(hue)


```python
tips.head()
```

```python
sns.relplot(x='total_bill',y = 'tip',data = tips,kind = 'scatter')
plt.show()
```

```python
# Specifying rows for subplots

sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           row ='sex' )
plt.show()
```

```python
# Specifying columns for subplots

sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
            col ='day'
           )
plt.show()
```

```python
# controlling subplots arrangement with col_wrap

sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           col ='day',
           col_wrap = 2
           )
plt.show()
```

```python
sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           col ='sex',
           row = 'smoker')
plt.show()
```

```python
# Ordering the rows
sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           col ='sex',
           row = 'smoker',
           row_order = ['No','Yes'])
plt.show()
```

```python
# Ordering the rows
sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           col ='sex',
           row = 'smoker',
           col_order = ['Female','Male'])
plt.show()
```

# Categorical plots


Categorical variables can be plotted alone to check show the count of components but reveals more when plotted against a numerical variable.


### Groups of Categorical data plots in Seaborn

Seaborn puts categorical plots in three main groups as follows:
1. Group that shows each observation
    * stripplot
    * swarmplot
2. Group thats shows abstract representations of the plots.
    * boxplot
    * Violiinplot
    * boxenplot
3. Group that shows statistical estimates
    * countplot
    * barplot
    * pointplot


### Categorical plot group that shows each observation

### stripplot

* Good for large data as compared to swarmplot

```python
# Create a figure and axes
fig, ax= plt.subplots()

sns.histplot(pokemon['special-defense'], ax=ax)

# Modify the properties of the plot
ax.set(xlabel="special-defense",
       xlim=(10,160),
       title="Distribution of special special defense")
# Add vertical lines for the median and mean
ax.axvline(x=pokemon['special-defense'].median(), color='red', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=pokemon['special-defense'].mean(), color='red', label='Mean', linestyle='-', linewidth=2)

# Display the plot
plt.show()
```
